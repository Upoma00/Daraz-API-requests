```python
import requests
data = []
for page in range(1,10):
    url= f'https://www.daraz.com.bd/smartphones/?ajax=true&page={page}&spm=a2a0e.home.cate_6.1.735212f7dWNMB4'
    response =  requests.get(url)
    data_json = response.json()
    print(f'{page} no page is done')
    all_phones = data_json['mods']['listItems']
    for item,phones in enumerate(all_phones):
        try:
            originalPrice = phones['originalPrice']
        except:
            originalPrice = phones['price']

        try:
            discount = phones['discount']
        except:
            discount = 'Not available'
        try:
            rating = phones['ratingScore']
        except:
            rating = 'Not available'
        d = {
            'name':phones['name'],
            'originalPrice': originalPrice,
            'price':phones['price'],
            'discount':discount,
            'ratingScore': rating,
            'review':phones['review'],
            'brandName':phones['brandName']

        }
        data.append(d)
```
